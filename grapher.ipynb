```python
def extract_accuracies(file_path):
    accuracies = []
    with open(file_path, 'r') as file:
        for line in file:
            if "Perturbed Accuracy" in line:
                accuracy = float(line.split(":")[1].strip().replace("%", ""))
                accuracies.append(accuracy)
    return accuracies

# Example usage:
import matplotlib.pyplot as plt

conceptizer_accuracies = extract_accuracies('282_results/progress_report/conceptizer.txt')
parameterizer_accuracies = extract_accuracies('282_results/progress_report/parametrizer.txt')
aggregator_accuracies = extract_accuracies('282_results/progress_report/aggregator.txt')

x_axis = list(range(0, 275, 25))

plt.plot(x_axis, aggregator_accuracies, label='Aggregator')
plt.plot(x_axis, conceptizer_accuracies, label='Conceptizer')
plt.plot(x_axis, parameterizer_accuracies, label='Parameterizer')

plt.xlabel('Epsilon (/255)')
plt.ylabel('Test Accuracy (%)')
plt.title('Accuracy vs FGSM Strength')
plt.legend()
plt.show()
# print(accuracies)
```
